```python
import pandas as pd

teams = pd.read_csv('data/team_batting_2025.csv')
print(teams.shape)
teams
```

```python
brewers_row = teams[teams['Tm'] == 'Milwaukee Brewers']
league_avg = teams[teams['Tm'] == 'League Average']

brewers_row
```

```python
compare_cols = ['R', 'HR', '2B', '3B', 'SB', 'BB', 'SO', 'BA', 'OBP', 'SLG']

comparison = pd.concat([
    brewers_row[compare_cols].reset_index(drop=True),
    league_avg[compare_cols].reset_index(drop=True)
])
comparison.index = ['Brewers', 'League Average']
comparison
```

Roster-Level Gap Analysis: Offesive Power

**Finding:** The Brewers offense is contact-oriented rather than power-oriented, which may be a roster weakness against strong pitching

**Supporting Data (2025 season, Brewers vs. league average):** 
- Batting Average: .258 vs .245 (+5%)
- On-Base Percentage: .332 vs .315 (+5%)
- Slugging Percentage: .403 vs .404 (essentially even)
- Home Runs: 166 vs 188 (below average)
- Strikeouts: 1266 vs 1355 (well below average)

**Interpretation:** Despite an above-average ability to get on base and make contact, Milwauke gets almiost no extra production from power. Their slugging is only league-average despite a higher batting average, suggesting they're compensating with doubles/triples rather than home runs. This profile can be exploited by strong pitching, where contact alone is less effective than raw power is producing runs
